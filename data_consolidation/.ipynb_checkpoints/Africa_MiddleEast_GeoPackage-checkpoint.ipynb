# <font color = 'darkblue'> Wells


```python
# Read the data
# =============================================
fpWells = "OGIM_AF_ME_v1a\\Wells"
all_Wells = glob.glob(fpWells + "\*.shp")

all_Wells_data = []
for file in all_Wells:
    gdf_ = gpd.read_file(file)
    # Check crs
    
    if "epsg:4326" in str(gdf_.crs):
        all_Wells_data.append(gdf_)
        
    else:
        dat_trans = gdf_.to_crs("epsg:4326")
        all_Wells_data.append(data_trans)
        
all_Wells_data2 = pd.concat(all_Wells_data).reset_index()

# Create unique OBJECT IDs
all_Wells_data2['OBJECT_ID'] = np.arange(1,len(all_Wells_data2.OBJECT_ID)+1)

# =============================================
```

```python
nulls_comp, gdf_nulls2 = check_null_geometries(all_Wells_data2, id_attr='OBJECT_ID')
```

```python
# OKAY, for NOW, drop `None` geometries
# =============================================
wells_data_v2 = all_Wells_data2[~all_Wells_data2.OBJECT_ID.isin(nulls_comp)]
nulls_comp2, gdf_nulls22 = check_null_geometries(wells_data_v2, id_attr='OBJECT_ID')
```

```python
# Exclude "category_new" in consolidated geopackage
wells_data_v3 = wells_data_v2[~wells_data_v2.CATEGORY.isin(['category_new'])]
print (wells_data_v3.shape[0])
```
